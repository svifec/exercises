```python
#  DataFrame (    df)
df = pd.read_excel('the number of road motor vehicles by province.xls')

# ,   Total    
df_clean = df[df['Provinces'] != 'Toplam-Total'].copy()

# 1. -10     
top_10_provinces = df_clean.nlargest(10, 'Total')

plt.figure(figsize=(12, 6))
plt.bar(top_10_provinces['Provinces'], top_10_provinces['Total'],color='#cffc92')
plt.title('-10      ')
plt.xlabel('')
plt.ylabel('  (.)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
df1 = pd.read_excel('the number of road motor vehicles by province.xls')
df2 = pd.read_excel('the number of road motor vehicles handed over by province.xls')

#    
total_row1 = df1[df1['Provinces'] == 'Toplam-Total'].iloc[0]
total_row2 = df2[df2['Provinces'] == 'Toplam-Total'].iloc[0]

# 
categories = ['Car', 'Minibus', 'Bus', 'Small truck', 'Truck', 'Motorcycle', 'purpose', 'Tractor']

values1 = [total_row1[col] for col in categories]
values2 = [total_row2[col] for col in categories]

#  
y = range(len(categories))
fig, ax = plt.subplots(figsize=(10, 8))

#  
ax.barh([i - 0.2 for i in y], values1, height=0.4, label=' 2', color='lightcoral')
ax.barh([i + 0.2 for i in y], values2, height=0.4, label=' 5', color='lightblue')

# 
ax.set_xlabel('   (.)')
ax.set_ylabel('  ')
ax.set_title('     ')
ax.set_yticks(y)
ax.set_yticklabels(categories)
ax.legend(loc='lower right')
ax.grid(axis='x', linestyle='', alpha=0.6)
ax.set_xscale('log')
if ax.get_xscale() == 'linear':
    ax.set_xlim(left=0)

plt.tight_layout()
plt.show()
```
