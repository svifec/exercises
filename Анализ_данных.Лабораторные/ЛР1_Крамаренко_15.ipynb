#### Задание 3

Сводная таблица (pivot_table) - среднее количество покупок для

Для этого используется метод pivot_table для создания сводной таблицы и sort_values со значением False для сортировки по убыванию, а также round(2) для округления.

Самое большое среднее значение 2.53 для уровня образования Master, а самое маленькое 1.40 для Basic.

### Дополнительное задание

Создать столбец - возраст (расчетный). Создать столбец “Категория
возраста (Year_Birth)” (с помощью категоризации). Выделить минимум 3
категории (молодой, старый, средний), фильтрацию для уровня возраста
выбрать самостоятельно, аргументировать выбор. Создать сводную
таблицу: средний и медианный возраст по Marital_Status и категории
возраста.


```python
# код выполнения задания
current_year = pd.Timestamp.now().year
df['Age'] = current_year - df['Year_Birth']

age_bins = [0, 35, 55, 100]
age_labels = ['Молодой', 'Средний', 'Пожилой']

df['Категория возраста (Year_Birth)'] = pd.cut(
    df['Age'],
    bins=age_bins,
    labels=age_labels,
    right=True
)

t=input('Введите категорию возраста ')
filtered_df = df[df['Категория возраста (Year_Birth)'] == t]
filtered_df.head(20)
```

Метод cut гарантирует, что все значения будут отнесены к соответствующим категориям согласно заданным границам, исключая субъективизм и ошибки ручной категоризации. Параметр right=True обеспечивает четкое определение принадлежности к интервалам. Выбор интерактивного подхода к фильтрации возрастных категорий через input()  позволяет гибко исследовать различные возрастные группы без необходимости модификации кода, что особенно ценно при работе с меняющимися исследовательскими данными.

```python
pivot_table = pd.pivot_table(
    df,
    values='Age',
    index=['Marital_Status', 'Категория возраста (Year_Birth)'],
    aggfunc=['mean', 'median'],
    margins=True,
    margins_name='Все категории',
    observed=False
)

pivot_table.columns = ['Средний возраст', 'Медианный возраст']
pivot_table.round(1)
```

Была создана сводная таблица - средний и медианный возраст по Marital_Status и категории возраста

Создать столбец “Категория дохода (Income)” (с помощью категоризации).
Выделить минимум 3 категории (высокий, низкий, средний), фильтрацию
для уровня дохода выбрать самостоятельно, аргументировать выбор.
Создать сводные таблицы: средний и медианный NumDealsPurchases по
категории дохода и Education.

```python
tercile_33 = df['Income'].quantile(0.33)
tercile_66 = df['Income'].quantile(0.66)
income_bins = [0, tercile_33, tercile_66, df['Income'].max() + 1]
income_labels = ['Низкий', 'Средний', 'Высокий']

df['Категория дохода (Income)'] = pd.cut(
    df['Income'],
    bins=income_bins,
    labels=income_labels,
    right=False
)
t=input('Введите категорию дохода ')
middle_income = df[df['Категория дохода (Income)'] == t]
middle_income.head(20)
```

Выбор интерактивного подхода к фильтрации уровня дохода через input()  позволяет гибко исследовать различный уровень дохода без необходимости модификации кода. Фильтрация основывается на статистически обоснованных категориях, созданных через терцильное разделение. Также был применен метод cut()

```python
pivot_table = pd.pivot_table(
    df,
    values='Number_of_Purchases',
    index=['Категория дохода (Income)', 'Education'],
    aggfunc=['mean', 'median'],
    margins=True,
    margins_name='Все категории',
    observed=False
)

pivot_table.columns = ['Средние покупки ', 'Медианные покупки ']
pivot_table.round(2)
```

Была создана сводная таблица с помощью метода pivot_table - средний и медианный Number_of_Purchases по категории дохода и Education
